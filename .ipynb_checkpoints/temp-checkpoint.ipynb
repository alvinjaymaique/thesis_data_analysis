Summary Table of Cluster Characteristics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the clustered data
df = pd.read_csv('clustered_anomalies.csv')

# Create summary stats by cluster
cluster_summary = df.groupby('anomaly_type').agg({
    'voltage': ['mean', 'std', 'min', 'max'],
    'current': ['mean', 'std', 'min', 'max'],
    'power': ['mean', 'std', 'min', 'max'],
    'powerFactor': ['mean', 'std', 'min', 'max'],
    'cluster': 'count'  # Count of samples in each cluster
}).rename(columns={'cluster': 'samples'})

# Calculate percentage of total observations
total_samples = len(df)
cluster_summary['percentage'] = (cluster_summary['samples']['count'] / total_samples * 100).round(1)

# Sort by frequency (descending)
cluster_summary = cluster_summary.sort_values(('samples', 'count'), ascending=False)

# Display as formatted table
print(cluster_summary.to_string())
```

2D Scatter Plot with PCA Projection

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select numerical features for projection
features = ['voltage', 'current', 'power', 'powerFactor', 
            'voltage_deviation', 'frequency_deviation', 'pf_deviation',
            'power_voltage_ratio', 'current_voltage_ratio']

# Standardize features
X = StandardScaler().fit_transform(df[features])

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create DataFrame with PCA results and cluster information
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['anomaly_type'] = df['anomaly_type'].values
pca_df['location'] = df['location'].values

# Create PCA visualization
plt.figure(figsize=(12, 10))
sns.scatterplot(x='PC1', y='PC2', hue='anomaly_type', data=pca_df, alpha=0.7)
plt.title('PCA Visualization of Power Quality Anomalies')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('plots/pca_anomalies_visualization.png', dpi=300)
plt.close()
```

Example Waveforms for Each Cluster Type

```python
# Select representative samples from each anomaly type
anomaly_types = df['anomaly_type'].unique()

# Create a figure with subplots for each anomaly type
plt.figure(figsize=(15, len(anomaly_types)*3))

for i, anomaly in enumerate(anomaly_types):
    # Get a sample of records for this anomaly type (sorted by timestamp)
    sample = df[df['anomaly_type']==anomaly].sort_values('timestamp').head(50)
    
    plt.subplot(len(anomaly_types), 1, i+1)
    plt.plot(sample['timestamp'], sample['voltage'], 'b-', label='Voltage')
    plt.title(f'Characteristic Pattern: {anomaly}')
    plt.ylabel('Voltage (V)')
    
    # Add a second axis for current
    ax2 = plt.twinx()
    ax2.plot(sample['timestamp'], sample['current'], 'r-', label='Current')
    ax2.set_ylabel('Current (A)', color='r')
    
    # Only show x-axis labels for the bottom plot
    if i == len(anomaly_types)-1:
        plt.xlabel('Time')
    
    # Add legend
    lines1, labels1 = plt.gca().get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.tight_layout()
plt.savefig('plots/anomaly_waveform_patterns.png', dpi=300)
plt.close()
```

Location-Based Analysis

```python
# Create a heatmap showing anomaly types by location
plt.figure(figsize=(14, 10))

# Create a crosstab of anomaly types vs locations
location_anomaly = pd.crosstab(df['location'], df['anomaly_type'], normalize='index') * 100

# Plot as heatmap
sns.heatmap(location_anomaly, cmap='YlOrRd', annot=True, fmt='.1f')
plt.title('Distribution of Anomaly Types by Location (%)')
plt.ylabel('Location')
plt.xlabel('Anomaly Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('plots/location_anomaly_heatmap.png', dpi=300)
plt.close()
```

Anomaly Frequency Over Time

```python
# Convert timestamp to datetime if it isn't already
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract date component
df['date'] = df['timestamp'].dt.date

# Count anomalies by type and date
anomaly_time_series = df.groupby(['date', 'anomaly_type']).size().unstack().fillna(0)

# Plot time series
plt.figure(figsize=(15, 8))
anomaly_time_series.plot(kind='line', ax=plt.gca())
plt.title('Anomaly Occurrences Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Occurrences')
plt.legend(title='Anomaly Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/anomaly_time_series.png', dpi=300)
plt.close()
```
