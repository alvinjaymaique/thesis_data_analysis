```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Make sure plots directory exists
os.makedirs('clusters', exist_ok=True)

# Load the data
df_with_clusters = pd.read_csv('clustered_anomalies.csv')
x_tick_size = 12
# Create a mapping dictionary using the more concise labels from the table
cluster_labels = {
    0: "LightLoad_VoltageSurge",
    1: "Idle_Overvoltage",
    2: "LightLoad_Undervoltage",
    3: "HighLoad_VoltageInstability",
    4: "HighLoad_SevereTransients",
    5: "Idle_Undervoltage",
    6: "LowPF_ReactiveLoad",
    7: "ModeratePF_MinorSurge",
    8: "HighLoad_MixedAnomalies",
    9: "LightLoad_Undervoltage_LowPF",
    10: "LightLoad_MinorSurge",
    11: "HighLoad_Optimal",
    12: "Idle_Stable",
    13: "HighLoad_Excellent",
    14: "PeakLoad_Excellent"
}

# Map cluster numbers to labels
df_with_clusters['cluster_label'] = df_with_clusters['cluster'].map(cluster_labels)

# 1. Voltage Characteristics by Cluster
plt.figure(figsize=(16, 10))
ax = sns.boxplot(x='cluster', y='voltage', data=df_with_clusters)
plt.axhline(y=230.0, color='green', linestyle='--', label='Nominal (230V)')
plt.axhline(y=242.6, color='red', linestyle='--', label='Surge Threshold')
plt.axhline(y=217.4, color='orange', linestyle='--', label='Dip Threshold')
plt.xlabel('Cluster Type', fontsize=12)
plt.ylabel('Voltage (V)', fontsize=12)
plt.title('Voltage Distribution by Cluster', fontsize=16)

# Use the label values instead of keys for x-ticks
cluster_numbers = sorted(cluster_labels.keys())
label_values = [cluster_labels[i] for i in cluster_numbers]
plt.xticks(ticks=range(len(cluster_labels)), labels=label_values, fontsize=x_tick_size, rotation=45, ha='right')

plt.legend(fontsize=10)
plt.grid(axis='y', alpha=0.3)

# Add cluster numbers at the top of the plot
for i, cluster_num in enumerate(cluster_numbers):
    ax.text(i, plt.ylim()[1]*0.98, f"C{cluster_num}", ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('clusters/voltage_distribution_by_cluster.png', dpi=300)
plt.close()

# 2. Power Factor by Cluster
plt.figure(figsize=(16, 10))
ax = sns.boxplot(x='cluster', y='powerFactor', data=df_with_clusters)
plt.axhline(y=0.85, color='green', linestyle='--', label='Acceptable PF')
plt.axhline(y=0.7, color='orange', linestyle='--', label='Low PF')
plt.axhline(y=0.5, color='red', linestyle='--', label='Very Poor PF')
plt.xlabel('Cluster Type', fontsize=12)
plt.ylabel('Power Factor', fontsize=12)
plt.title('Power Factor Distribution by Cluster', fontsize=16)
plt.xticks(ticks=range(len(cluster_labels)), labels=label_values, fontsize=x_tick_size, rotation=45, ha='right')
plt.legend(fontsize=10)
plt.grid(axis='y', alpha=0.3)

# Add cluster numbers at the top
for i, cluster_num in enumerate(cluster_numbers):
    ax.text(i, plt.ylim()[1]*0.98, f"C{cluster_num}", ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('clusters/power_factor_distribution_by_cluster.png', dpi=300)
plt.close()

# 3. Frequency by Cluster
plt.figure(figsize=(16, 10))
ax = sns.boxplot(x='cluster', y='frequency', data=df_with_clusters)
plt.axhline(y=60.0, color='green', linestyle='--', label='Nominal (60Hz)')
plt.axhline(y=59.2, color='orange', linestyle='--', label='Low Threshold')
plt.axhline(y=60.8, color='orange', linestyle='--', label='High Threshold')
plt.xlabel('Cluster Type', fontsize=12)
plt.ylabel('Frequency (Hz)', fontsize=12)
plt.title('Frequency Distribution by Cluster', fontsize=16)
plt.xticks(ticks=range(len(cluster_labels)), labels=label_values, fontsize=x_tick_size, rotation=45, ha='right')
plt.legend(fontsize=10)
plt.grid(axis='y', alpha=0.3)

# Add cluster numbers at the top
for i, cluster_num in enumerate(cluster_numbers):
    ax.text(i, plt.ylim()[1]*0.98, f"C{cluster_num}", ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('clusters/frequency_distribution_by_cluster.png', dpi=300)
plt.close()

# 4. Power Consumption by Cluster (log scale)
plt.figure(figsize=(16, 10))
ax = sns.boxplot(x='cluster', y='power', data=df_with_clusters)
plt.yscale('log')
plt.xlabel('Cluster Type', fontsize=12)
plt.ylabel('Power (W) - Log Scale', fontsize=12)
plt.title('Power Consumption Distribution by Cluster', fontsize=16)
plt.xticks(ticks=range(len(cluster_labels)), labels=label_values, fontsize=x_tick_size, rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Add cluster numbers at the top
for i, cluster_num in enumerate(cluster_numbers):
    ax.text(i, plt.ylim()[1]*0.95, f"C{cluster_num}", ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('clusters/power_consumption_distribution_by_cluster.png', dpi=300)
plt.close()

# 5. Current by Cluster (log scale)
plt.figure(figsize=(16, 10))
ax = sns.boxplot(x='cluster', y='current', data=df_with_clusters)
plt.yscale('log')
plt.xlabel('Cluster Type', fontsize=12)
plt.ylabel('Current (A) - Log Scale', fontsize=12)
plt.title('Current Distribution by Cluster', fontsize=16)
plt.xticks(ticks=range(len(cluster_labels)), labels=label_values, fontsize=x_tick_size, rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Add cluster numbers at the top
for i, cluster_num in enumerate(cluster_numbers):
    ax.text(i, plt.ylim()[1]*0.95, f"C{cluster_num}", ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('clusters/current_distribution_by_cluster.png', dpi=300)
plt.close()

# 6. Cluster Size Distribution
plt.figure(figsize=(16, 10))
cluster_counts = df_with_clusters['cluster'].value_counts().sort_index()
x_pos = np.arange(len(cluster_counts))
bars = plt.bar(x_pos, cluster_counts.values, color='skyblue', edgecolor='black')

# Add value labels above each bar
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
             f'{bar.get_height()}', ha='center', fontsize=10)

plt.xlabel('Cluster Type', fontsize=12)
plt.ylabel('Number of Samples', fontsize=12)
plt.title('Cluster Size Distribution', fontsize=16)
plt.xticks(x_pos, [cluster_labels[i] for i in cluster_counts.index], fontsize=x_tick_size, rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('clusters/cluster_size_distribution.png', dpi=300)
plt.close()

print("Individual visualizations created and saved to 'clusters/' directory")
```
